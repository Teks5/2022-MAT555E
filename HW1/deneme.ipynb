
# Question 1 Solution 

```python
import numpy as np
def  Riemann(a,b,N,fn):            # middle point riemann is used.
    lower_value = min(a,b)
    upper_value = max(a,b)
    step_size = (upper_value-lower_value)/N
    R_sum = 0
    for i in range (N):
        R_sum = R_sum+fn(((i*step_size)+((i+1)*step_size))/2)
    return R_sum*step_size
    
```

# Question 2 Solution

```python
import numpy as np
import random
import matplotlib.pyplot as plt
def fu(N):
    pts = np.zeros((1,2))
    cntr=0
    while cntr<N:
        circle_point1 = random.uniform(-1,1)
        root = np.sqrt(1-(circle_point1)**2)
        circle_point2 = np.random.choice([root,-root])
        circle_points = np.array([circle_point1,circle_point2])
        circle_points = np.resize(circle_points,(1,2))
        pts = np.concatenate((pts,circle_points))
        cntr=cntr+1
    return(pts[1:,:])

result = fu(50)

figure, axes = plt.subplots(figsize=(10,10)) 
axes.scatter(result[:,0],result[:,1],color='r')
cc = plt.Circle(( 0 , 0 ), 1 , alpha=0.1 ,color='black') 
axes.set_aspect( 1 ) 
axes.add_artist( cc )
axes.set_xlim(-1.1, 1.1)
axes.set_ylim(-1.1, 1.1)
plt.show()
```

# Question 3 Solution 

```python
def fu_3d(N):
    pts = np.zeros((1,3))
    ptso = np.zeros((1,3))
    cntr=0
    while cntr<N:
        rnd_pts = np.random.randn(1,3)
        if (rnd_pts[0,0]**2+rnd_pts[0,1]**2+rnd_pts[0,2]**2) <= 1:
            pts=np.concatenate((pts,rnd_pts))
            cntr=cntr+1
        else:
            ptso=np.concatenate((ptso,rnd_pts))
    return(pts[1:,:],ptso[1:,:]) 
```

```python
result = fu_3d(10)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
# draw sphere
u, v = np.mgrid[0:2*np.pi:50j, 0:np.pi:50j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
# alpha controls opacity
ax.plot_surface(x, y, z, color="g", alpha=0.3)
ax.scatter(result[0][:,0],result[0][:,1],result[0][:,2],color='r')
ax.scatter(result[1][:,0],result[1][:,1],result[1][:,2],color='b')
ax.set_xlim([-1.5,1.5])
ax.set_ylim([-1.5,1.5])
ax.set_zlim([-1.5,1.5])
```

# Question 4 

```python
import numpy as np
def fu_nd(n_dim,N):     # two parameters input, n_dim = the dimension of the demanded list, N = the demanded number of the point inside the n-dim unit array
    pts = np.zeros((1,n_dim))
    ptso = np.zeros((1,n_dim))
    cntr=0
    while cntr<N:
        rnd_pts = np.random.randn(1,n_dim)
        
        if (np.sum(rnd_pts**2)) <= 1:    # sum of the squares of all elements. If it is below zero it is added to the list.
            pts=np.concatenate((pts,rnd_pts))
            cntr=cntr+1
        else:
            ptso=np.concatenate((ptso,rnd_pts))
    return(pts[1:,:],ptso[1:,:]) 
```

```python
results = fu_nd(6,10)
inside = results[0]
outside = results[1]
print('Inside of the 1',inside)
print('Outside of the 1',outside)
```
