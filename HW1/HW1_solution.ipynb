
# Question 1 Solution 

```python
import numpy as np
def  Riemann(a,b,N,fn):            # middle point riemann is used.
    lower_value = min(a,b)
    upper_value = max(a,b)
    step_size = (upper_value-lower_value)/N
    R_sum = 0
    for i in range (N):
        R_sum = R_sum+fn(((i*step_size)+((i+1)*step_size))/2)
    return R_sum*step_size
    
```

# Question 2 Solution

```python
import numpy as np
import random
import matplotlib.pyplot as plt
def fu(N):
    pts = np.zeros((1,2))
    cntr=0
    while cntr<N:
        circle_point1 = random.uniform(-1,1)
        root = np.sqrt(1-(circle_point1)**2)
        circle_point2 = np.random.choice([root,-root])
        circle_points = np.array([circle_point1,circle_point2])
        circle_points = np.resize(circle_points,(1,2))
        pts = np.concatenate((pts,circle_points))
        cntr=cntr+1
    return(pts[1:,:])

result = fu(50)

figure, axes = plt.subplots(figsize=(10,10)) 
axes.scatter(result[:,0],result[:,1],color='r')
cc = plt.Circle(( 0 , 0 ), 1 , alpha=0.1 ,color='black') 
axes.set_aspect( 1 ) 
axes.add_artist( cc )
axes.set_xlim(-1.1, 1.1)
axes.set_ylim(-1.1, 1.1)
plt.show()
```

# Question 3 Solution 

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def fu_3d(N):
    pts = np.zeros((1,3))
    cntr=0
    while cntr<N:
        circle_point1 = random.uniform(-1,1)
        root = np.sqrt(1-(circle_point1)**2)         # y^2 + z^2 = 1 - x^2
        circle_point2 = random.uniform(-root,root)
        root2 = np.sqrt((root)**2-(circle_point2)**2)        # z^2 = 1-x^2-y^2
        circle_point3 = np.random.choice([-root2,root2])
        circle_points = np.array([circle_point1,circle_point2,circle_point3])
        circle_points = np.resize(circle_points,(1,3))
        pts = np.concatenate((pts,circle_points))
        cntr=cntr+1
    return(pts[1:,:]) 

result = fu_3d(10)


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
# draw sphere
u, v = np.mgrid[0:2*np.pi:50j, 0:np.pi:50j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
# alpha controls opacity
ax.plot_surface(x, y, z, color="black", alpha=0.1)
ax.scatter(result[:,0],result[:,1],result[:,2],color='r')
ax.set_xlim([-1.5,1.5])
ax.set_ylim([-1.5,1.5])
ax.set_zlim([-1.5,1.5])
```

# Question 4 

```python
import numpy as np
def fu_nd(n_dim,N):     # two parameters input, n_dim = the dimension of the demanded list, N = the demanded number of the point inside the n-dim unit array
    pts = np.zeros((1,n_dim))
    ptsn = np.zeros((1,n_dim))
    cntr=0
    while cntr<N:
        point1 = random.uniform(-1,1)     # first point choose random uniformly.
        root = np.sqrt(1-(point1)**2)
        ptsn[0,0] = point1
        for i in range (1,n_dim-1):
            new_point = random.uniform(-root,root)     # other points are created the remained sum.
            root = np.sqrt((root)**2-(new_point)**2)    
            ptsn[0,i]= new_point
        ptsn[0,-1]=np.random.choice([-root,root])    # last point is decided with remained root.
        pts=np.concatenate((pts,ptsn))
        cntr=cntr+1
    return(pts[1:,:]) 

result = fu_nd(4,10)   # This part can be changed to obtain different result.
sum(result[7]**2)     #  Sum of a random row.
```
